```python
#Uniform distribution
n_iter=100000
n_bin=round(n_iter/100)

arr=np.empty(n_iter)

for i in range(n_iter):
    arr[i]=random.random()
#np.histogram provides two arrays
n, bins, patches = plt.hist(arr, n_bin, range=(0,1))

```

```python
# distribtion
n_iter=100000
n_bin=round(n_iter/100)

arr=np.empty(n_iter)

for i in range(n_iter):
    arr[i]=random.random()+random.random()
#np.histogram provides two arrays
n, bins, patches = plt.hist(arr, n_bin, range=(0,2))

##This is not a normal distribution
```

```python
#Normal distribtion
n_iter=100000
n_bin=round(n_iter/100)

mu=0
sigma=1

arr=np.random.normal(mu,sigma,n_iter)

#np.histogram provides two arrays
n, bins, patches = plt.hist(arr, n_bin, range=(-5,5))
```

```python
#Normal distribtion
n_iter=100000
n_bin=round(n_iter/100)

mu=0
sigma=1

arr=np.random.normal(mu,sigma,n_iter)
arr2=np.random.normal(mu,sigma,n_iter)

#np.histogram provides two arrays
n, bins, patches = plt.hist(arr+arr2, n_bin, range=(-5,5))

##We can see that the sigma has been added due to the addition of two numbers from normal distributions
```
